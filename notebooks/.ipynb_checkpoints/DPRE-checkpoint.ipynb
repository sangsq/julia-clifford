```julia
using LinearAlgebra
using Statistics
using PyPlot
```

```julia
function rand_potential(nx, ny)
    rg = 1:3
    return rand(rg, ny, nx)
end

function dpre(y0, V)
    m, n = size(V)
    infty = 2^32-1
    Jy = 1
    s = fill(infty, m, n)
    s_diff = fill(0, m, n)

    s[y0, 1] = V[y0, 1]
    
    for i in 2:n
        for j in 1:m
            l = argmin([s[k, i-1] + abs(k-j) for k in 1:m])
            s[j, i] = s[l, i-1] + abs(l-j) + V[j, i]
        end
    end
    
    return s
end

function dpre2(y0, V)
    m, n = size(V)
    infty = 2^32-1
    Jy = 1
    E = fill(infty, m, n)
    Ep = fill(infty, m, n)
    Epp = fill(infty, m, n)
    
    E[y0, 1] = V[y0, 1]
    Ep[y0, 1] = 0
    Epp[y0, 1] = 0
    
    for i in 2:n
        Epp[1, i] = E[1, i-1]
        for j in 1:1:m-1
            Epp[j+1, i] = min(E[j+1, i-1], Epp[j, i] + Jy)
        end
        Ep[m, i] = Epp[m, i]
        for j in (m-1):(-1):1
            Ep[j, i] = min(Epp[j, i], Ep[j+1, i] + Jy)
        end
        for j in 1:m
            E[j, i] = Ep[j, i] + V[j, i]
        end
    end
    return E
end
            


# function shortest_dpath(y0, V)
#     m, n = size(V)
#     infty = 2^32
#     b = 0
#     s = fill(infty, m)
#     s_tmp = fill(infty, m)
#     s_top = fill(0, n)
    
#     s[y0] = V[y0]
    
#     for i in 1:n
#         while true
#             s_tmp .= s
#             no_change = true
#             for j in 1:m
#                 tmp = min(
#                     s_tmp[j], 
#                     j==1 ? infty : s_tmp[j-1] + V[j, i] + b, 
#                     j==m ? infty : s_tmp[j+1] + V[j, i] + b, 
#                 )
#                 if tmp != s[j]
#                     s[j] = tmp
#                     no_change = false
#                 end
#             end
#             if no_change
#                 break
#             end
#         end
        
#         s_top[i] = s[y0]
        
#         if i==n
#             return s, s_top
#         end
        
#         for j in 1:m
#             s[j] += V[j, i+1]
#         end
#     end
# end


function dpre_f_rms(m, n, n_sample)
    fs = [minimum(dpre(1, rand_potential(m, n)), dims=1)[1,:] for _ in 1:n_sample]
    return std(fs)
end

function dpre_z_rms(m, n, n_sample)
    fs = [[a[1] for a in argmin(dpre(1, rand_potential(m, n)), dims=1)[1, :]] for _ in 1:n_sample]
    return std(fs)
end
    


function MI_sim(n_step, step_size, A_list, n_average)
    n = maximum(A_list) * 2
    result = zeros(n_step, length(A_list))
    for _ in 1:n_average
        V = rand_potential(n, n_step * step_size)
        for t in 1:n_step
            V_tmp = view(V, 1 : t*step_size, :)
            s, s_top = shortest_dpath(1, V_tmp)
            
            result[t, :] += 2 .* s_top[A_list] - s_top[2 .* A_list]
        end
    end
    result ./= n_average
    return result
end
```
