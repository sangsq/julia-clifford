```julia
function mized_op!(state, i, j, channel, p, dilation)
    _, n = size(state)
    if rand() < p && rand() < dilation
        channel(state, i)
    else
        if rand() < dilation
            cl = random_clifford(2)
            clifford_action!(cl, state, [i, j])
        end
    end
    return nothing
end

function adam_op!(state, i, j, channel, p, dilation)
    _, n = size(state)
    if rand() < dilation
        cl = random_clifford(2)
        clifford_action!(cl, state, Int[i, j])
    end
    if rand() < p && rand() < dilation
        channel(state, i)
    end
    if rand() < p && rand() < dilation
        channel(state, j)
    end
    return nothing
end

function bw_update!(state, channel, p, op, dilation, periodic=true)
    _, n = size(state)
    for ini in [1,2]
        for i in ini : 2 : (periodic ? n : n-1)
            j = i==n ? 1 : i+1
            op(state, i, j, channel, p, dilation)
        end
    end
    return nothing
end
```

```julia
n = 64
p_list = 0.05:0.05:0.2
dilation = 0.01
n_average = 1000
n_step = 512
channel = depolarize!
op = adam_op!
ees = zeros(length(p_list), n_step)
ls = zeros(length(p_list), n_step)

for i in 1:length(p_list)
    p = p_list[i]
    for _ in 1:n_average
        state = all_up(n)
        for k in 1:n_step
            ees[i, k] += entropy(state)
            ls[i, k] += mean_stab_length(state)
            bw_update!(state, channel, p, op, dilation)
        end
    end
end
ees ./= n_average
ls ./= n_average
result = ees, ls

@save "./cache/dp_slow.jld2" result
```

```julia
function tmppp()
    n = 64
    p_list = 0.1:0.1:0.5
    dilation = 0.2
    n_average = 1e5
    n_step = 60
    channel = depolarize!
    op = adam_op!

    ees = zeros(length(p_list), n_step)
    ls = zeros(length(p_list), n_step)
    nc = zeros(length(p_list), n_step)

    for i in 1:length(p_list)
        p = p_list[i]
        nn_step = Int(floor(n_step / p^0.5 * n_step))
        for _ in 1:n_average
            state = all_up(n)
            for k in 1:n_step
                
                ees[i, k] += entropy(state)
                ep = end_points(state)
                ls[i, k] += sum(ep[:,2] - ep[:, 1])
                nc[i, k] += size(ep, 1)
                if ees[i, k] == n
                    ees[i, k+1:end] .+= n
                    break
                end
                
                bw_update!(state, channel, p, op, dilation)
            end
        end
    end
    ees ./= n_average
    result = ees, ls ./ nc
    return result
end

result = tmppp()

@save "./cache/dp_quick_long.jld2" result
```

```julia
function sim_lin(p, q=1, d=0.001)
    i = 1
    t = 0
    while true
        t += 1
        if rand() < d
            i += 1
        end
        if rand() < d * q
            if rand() < 0.5
                i -= 1
            else
                i += 1
            end
        end
        if rand() < 1 - (1 - d * p) ^ i
            return t
        end
    end
end

q = 1
d = 0.01
n_average = 10000
result = []
p_list = 0.1:0.1:0.8

for p in p_list
    push!(result, mean([sim_lin(p, q, d) for _ in 1:n_average]))
end

plot(p_list, result, "-o")
plot(p_list, 100 .* p_list .^ (-0.5), "-o")
xscale("log")
yscale("log")
```

```julia
function life_time_sim(p, q)
    s = 0
    t = 0
    while true
        t += 1
        s += 1
        if rand() < q
            if rand() < 0.5
                s += 1
            else
                s = max(1, s-1)
            end
        end
        if rand() < 1 - (1 - p) ^ s
            return s, t
        end
    end
end
```

```julia
p_list = 0.01:0.01:0.1
q = 0.3
max_len = 1000
count = 1e7
result = zeros(Int, length(p_list), max_len)

len_by_t = zeros(Int, length(p_list), max_len)
for i in 1:length(p_list)
    p = p_list[i]
    for _ in 1:count
        s, t = life_time_sim(p, q)
        result[i, t] += 1
    end
end
```

```julia
r = 1 .- cumsum(result, dims=2) ./ count
fg, ax = subplots()
for i in 1:length(p_list)
    p = p_list[i]
    nv = 0.5
    ax.plot((1:max_len) .* p^nv, r[i, :], "-o", ms=2, label=p)
end

ax.set_xlim(-1, 7.5)
ax.set_yscale("log")
# ax.set_xscale("log")
```

```julia
function spread!(a, x, q)
    n = length(x)
    for i in 2:n-1
        a[i] = (1-q) * x[i] + q * x[i-1]
    end
    a[1] = (1 - q) * x[1]
    return nothing
end

function decay!(x, p)
    n = length(x)
    for i in 1:n
        x[i] *= (1 - p) ^ ((i-1)^(1-p))
    end
    return nothing
end
```

```julia
n = 500
n_step = 500
p_list = 0.01:0.01:0.1
q = 0.2

mat_list = [zeros(n_step, n) for _ in p_list]
[mat[1, 1] = 1 for mat in mat_list]

for (i, p) in enumerate(p_list)
    mat = mat_list[i]
    @views for t in 1:n_step-1
        spread!(mat[t+1, :], mat[t, :], q)
        decay!(mat[t+1, :], p)
    end
end
```

```julia
@views function operator_update!(state, dilation)
    n = div(length(state), 2)
    
    for i in 1:2:n-1
        if state[2i-1]==false && state[2i]==false && state[2i+1]==false && state[2i+2]==false
            continue
        end
        tmp = rand() < dilation
        if tmp
            rand!(state[2i-1: 2i+2])
            while state[2i-1]==false && state[2i]==false && state[2i+1]==false && state[2i+2]==false
                rand!(state[2i-1: 2i+2])
            end
        end
    end
    
    for i in 2:2:n-1
        if state[2i-1]==false && state[2i]==false && state[2i+1]==false && state[2i+2]==false
            continue
        end
        tmp = rand() < dilation
        if tmp
            rand!(state[2i-1: 2i+2])
            while state[2i-1]==false && state[2i]==false && state[2i+1]==false && state[2i+2]==false
                rand!(state[2i-1: 2i+2])
            end
        end
    end
    return nothing
end

function get_len_and_sp(state)
    l = div(findfirst(state)+1, 2)
    r = div(findlast(state)+1, 2)
    c = 0
    for i in l:r
        c += state[2i-1] || state[2i]
    end
    return r-l+1, c
end

function op_sim(n, p, dilation, n_step, n_average, max_count)
    state = zeros(Bool, 2n)
    lens_sum = zeros(Int, n_step)
    sps_sum = zeros(Int, n_step)
    counts = zeros(Int, n_step)
    
    for _ in 1:n_average
        state .= false
        state[n] = true
        for t in 1:n_step
            operator_update!(state, dilation)
            len, sp = 0, 0
            if counts[t] < max_count
                len, sp = get_len_and_sp(state)
                lens_sum[t] += len
                sps_sum[t]  += sp
                counts[t] += 1
            end
            if rand() < dilation && rand() < (1 - (1 - p) ^ sp)
                break
            end
        end
    end
    return lens_sum, sps_sum, counts
end


function op_sim2(n, p, dilation, n_step, n_popu)
    states = zeros(Bool, n_popu, 2n)
    is_suv = zeros(Bool, n_popu)
    suv_rate = zeros(n_step)
    mean_len = zeros(n_step)
    mean_sp = zeros(n_step)
    states[:, n] .= true
    
    @views for t in 1:n_step
        len_sum = 0
        sp_sum = 0
        suv_idx = Int[]
        is_suv .= false
    
        for i in 1:n_popu
            operator_update!(states[i, :], dilation)
            len, sp = get_len_and_sp(states[i, :])
            suv = rand() > dilation || rand() < (1 - p) ^ sp
            if suv
                is_suv[i] = true
                push!(suv_idx, i)
                sp_sum += sp
                len_sum += len
            end
        end
        
        
        n_suv = length(suv_idx)
        
        if n_suv==0
            mean_len[t:end] .= 0
            mean_sp[t:end] .= 0
            suv_rate[t:end] .= 0
            break
        end
        
        mean_len[t] = len_sum / n_suv
        mean_sp[t] = sp_sum / n_suv
        suv_rate[t] =  n_suv / n_popu
        
        for i in 1:n_popu
            if !is_suv[i]
                tmp = rand(suv_idx)
                states[i, :] .= states[tmp, :]
            end
        end
    end
    return suv_rate, mean_len, mean_sp
end       
```

```julia
n = 100
p_list = 0.02:0.02:0.1
dilation = 1
n_step = 45
n_average = 1e7


max_count = 1e10


k = length(p_list)
lens_sums = zeros(Int, k, n_step)
sps_sums = zeros(Int, k, n_step)
countss = zeros(Int, k, n_step)

for (i, p) in enumerate(p_list)
    lens_sum, sps_sum, counts = op_sim(n, p, dilation, n_step, n_average, max_count)
    lens_sums[i, :] = lens_sum
    sps_sums[i, :] = sps_sum
    countss[i, :] = counts
end
```

```julia
fg, ax = subplots(3, figsize=(5, 12))
for (i, p) in enumerate(p_list)
    x = 1:n_step
    y1 = lens_sums[i, :] ./ countss[i, :]
    y2 = sps_sums[i, :] ./ countss[i, :]
    y3 = countss[i, :]
    ax[1].plot(x, y1, "-o", ms=2, label=p)
    ax[2].plot(x, y2, "-o", ms=2, label=p)
    ax[3].plot(x, y3, "-o", ms=2, label=p)
end
ax[3].set_yscale("log")
```

```julia
u(x) = Int(floor(x))

function spread!(a, x, q)
    n = length(x)
    
    for i in 1:(n-u(sqrt(n))-1)
        a[i+u(q*sqrt(i))+1] += 0.5 * x[i]
        a[i-u(q*sqrt(i))+1] += 0.5 * x[i]
    end
    
    return nothing
end

function decay!(x, p)
    n = length(x)
    for i in 1:n
        x[i] *= (1 - p) ^ ((i-1)^(1-p))
    end
    return nothing
end
```

```julia
n = 500
n_step = 1000
p_list = 0.01:0.01:0.1
q = 0.8

mat_list = [zeros(n_step, n) for _ in p_list]
[mat[1, 1] = 1 for mat in mat_list]

for (i, p) in enumerate(p_list)
    mat = mat_list[i]
    @views for t in 1:n_step-1
        spread!(mat[t+1, :], mat[t, :], q)
        decay!(mat[t+1, :], p)
    end
end
```
