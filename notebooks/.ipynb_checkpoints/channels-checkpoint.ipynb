```julia
include("../channels.jl")
using PyPlot
using LaTeXStrings
using Profile
```

```julia
function mized_op!(state, i, j, channel, p, dilation)
    _, n = size(state)
    if rand() < p && rand() < dilation
        channel(state, i)
    else
        if rand() < dilation
            cl = random_clifford(2)
            clifford_action!(cl, state, [i, j])
        end
    end
    return nothing
end

function adam_op!(state, i, j, channel, p, dilation)
    _, n = size(state)
    if rand() < dilation
        cl = random_clifford(2)
        clifford_action!(cl, state, Int[i, j])
    end
    if rand() < p && rand() < dilation
        channel(state, i)
    end
    if rand() < p && rand() < dilation
        channel(state, j)
    end
    return nothing
end

function bw_update!(state, channel, p, op, dilation, periodic=true)
    _, n = size(state)
    for i in 1 : (periodic ? n : n-1)
        j = i==n ? 1 : i+1
        op(state, i, j, channel, p, dilation)
    end
    return nothing
end
```

### Depolarizing channel

```julia
n = 64
p_list = 0.05:0.05:0.2
dilation = 0.01
n_average = 1000
n_step = 512
channel = depolarize!
op = adam_op!
ees = zeros(length(p_list), n_step)

for i in 1:length(p_list)
    p = p_list[i]
    for _ in 1:n_average
        state = all_up(n)
        for k in 1:n_step
            ees[i, k] += entropy(state)
            bw_update!(state, channel, p, op, dilation)
        end
    end
end
ees ./= n_average
```

```julia
for i in 1:length(p_list)
    x = (0:n_step-1)
    plot(x, ees[i, :], ".-", label=string("p=", p_list[i]))
end
xlabel(L"$t$")
ylabel("state entropy")
legend()
```

```julia
nv = 1
for i in 1:length(p_list)
    x = (0:n_step-1)
    plot(x .* p_list[i].^nv, ees[i, :], ms=1, ".-", label=string("p=", p_list[i]))
end
xlabel(L"$t \times p$")
ylabel(L"$S$")
legend()
```

```julia
nv = 0.75
for i in 1:length(p_list)
    x = (0:n_step-1)
    plot(x .* p_list[i].^nv, ees[i, :], ms=1, ".-", label=string("p=", p_list[i]))
end
xlabel(L"$t \times p$")
ylabel(L"$S$")
legend()
```

```julia
n = 64
cut = 32
p_list = 0.02:0.02:0.1
n_average = 500
n_step = 20
channel = depolarize!
op = adam_op!
mis = zeros(length(p_list), n_step)

for i in 1:length(p_list)
    p = p_list[i]
    for _ in 1:n_average
        state = all_up(n)
        for k in 1:n_step
            mis[i, k] += left_ee_on_all_cuts(state)[cut+1] + right_ee_on_all_cuts(state)[cut+1] - entropy(state)
            bw_update!(state, channel, p, op)
        end
    end
end
mis ./= n_average

for i in 1:length(p_list)
    x = (0:n_step-1)
    plot(x, mis[i, :], ".-", label=string("p=", p_list[i]))
end

xlabel(L"$t \times p^\nu$")
ylabel(L"$I(L/2:L/2)$")
legend() 
```

```julia
nu = 0.75
for i in 1:length(p_list)
    x = (0:n_step-1)
    p = p_list[i]
    plot(x .* p.^nu, mis[i, :] .* p, ".-", label=string("p=", p))
end
xlabel(L"$t \times p^\nu$")
ylabel(L"$I(L/2:L/2) \times p$")
legend() 
```
