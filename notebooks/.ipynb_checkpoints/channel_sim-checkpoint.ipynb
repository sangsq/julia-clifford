## Case 1
Here we consider an evolving fast scrambling channel that is interpersed with single qubit dephase/depolarize/damp channel. 
At each timestep,
1. All qubits are scrambled by a random n-qubit Clifford unitary
2. One qubit is completely depolarized
We check the capacity of this channel as a function of time.

```julia
function sim1(n, n_step, n_average, op)
    result = zeros(n_step, 3)
    for _ in 1:n_average
        ch = identity_channel(n)
        for t in 1:n_step
            result[t, :] .+= channel_decompose(ch)
            scramble!(ch)
            op(ch, 1)
        end
    end
    return result ./ n_average
end
```

```julia
n = 32
n_step = 64
n_average = 100
op = z_dephase!

result_uq2o3 = sim1(n, n_step, n_average, op);
```

```julia
result = result_uq2o3
fg,ax = subplots(1)

x = 0:n_step-1
ax.plot(x, result[:, 1], label="discard")
ax.plot(x, result[:, 2], label="dephase")
ax.plot(x, result[:, 3], label="identity")

ax.legend()
```

```julia
n = 32
n_step = 64
n_average = 100
op = depolarize!olarize!

result_23ofi = sim1(n, n_step, n_average, op);
```

```julia
result = result_23ofi
fg,ax = subplots(1)

x = 0:n_step-1
ax.plot(x, result[:, 1], label="discard")
ax.plot(x, result[:, 2], label="dephase")
ax.plot(x, result[:, 3], label="identity")

ax.legend()
```

```julia
n = 32
n_step = 64
n_average = 100
op = z_damp!

result_v23re = sim1(n, n_step, n_average, op);
```

```julia
result = result_v23re
fg,ax = subplots(1)

x = 0:n_step-1
ax.plot(x, result[:, 1], label="discard")
ax.plot(x, result[:, 2], label="dephase")
ax.plot(x, result[:, 3], label="identity")

ax.legend()
```

# Case 2
Consider the following setting:
1. At the begining, there are $pn$ qubits
2. Augment the system with $(1-p)n$ qubits prepared in up state
3. Scramble all $n$ qubits
4. Measure out (discard) $qn$ qubits

```julia
function sim2(n, p, q, n_average, n_step)
    result = zeros(n_step)
    for _ in 1:n_average
        ch = identity_channel(Int(p*n))
        add_qubits!(ch, Int((1-p)*n))
        for t in 1:n_step
            result[t] += channel_decompose(ch)[3]
            scramble!(ch)
            for i in 1:Int(q*n)
                depolarize!(ch, i)
            end
        end
    end
    result ./= n_average
    return result
end
```

```julia
p = 0.5
q_list = 0:0.125:1
n_list = [16, 32, 64, 128]
n_average = 10
result = zeros(length(n_list), length(q_list))

for (i, n) in enumerate(n_list)
    for (j, q) in enumerate(q_list)  
        result[i, j] = sim2(n, p, q, n_average, 2)[end]
    end
end

result_6vqa2 = result;
```

```julia
result = result_6vqa2
fg,ax = subplots(1)

x = q_list
for (i, n) in enumerate(n_list)
    ax.plot(x, result[i,:]./n, label=n)
end
ax.legend()

# ax.legend()
```
