```julia
include("../cliff7.jl")
using PyPlot
```

```julia
function zzx_layer!(state)
    m, n = size(state)
    X = (0, Bool[1, 0])
    ZZ = (0, Bool[0, 1, 0, 1])
    for i in 1:n
        j = i==n ? 1 : i+1
        if rand() < 0.5
            fps_measurement!(state, ZZ, [i, j])
        end
    end
    for i in 1:n
        if rand() < 0.5
            fps_measurement!(state, X, [i])
        end
    end
    return nothing
end



function zzx_mi_sim(n, n_step)
    ## let n=8k for some k

    mis = zeros(div(n, 4))
    burnin = 2n
    step_size = div(n, 8)

    state = all_plus(n)
    for _ in 1:burnin
        zzx_layer!(state)
    end
    for _ in 1:n_step
        mis += ap_mutual_info(state, 0, div(n, 2), div(n, 2))[2:2:end]
        for _ in 1:step_size
            zzx_layer!(state)
        end
    end

    mis ./= n_step

    xs = 2:2:div(n, 2)
    eta = sin.((pi/n) .* xs) .^ 2
    return eta, mis
end
```

```julia
n1 = 256
n2 = 128
n_step = 1000

eta1, mis1 = zzx_mi_sim(n1, n_step)
eta2, mis2 = zzx_mi_sim(n2, n_step)
```

```julia
fg, ax = subplots(1)
ax.plot(eta1, mis1, "-o", ms=2, label=n1)
ax.plot(eta2, mis2, "-o", ms=2, label=n2)

ax.plot(eta1, eta1.^(1/3))

ax.set_xscale("log")
ax.set_yscale("log")
```
